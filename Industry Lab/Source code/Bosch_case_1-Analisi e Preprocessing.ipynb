#Analisi del Dataset e Preprocessing

Caricamento del dataset

```
parte1 = pd.read_excel("/gdrive/My Drive/Progetto Industry Lab/Bosch_case_1.1.xlsx")
parte2 = pd.read_excel("/gdrive/My Drive/Progetto Industry Lab/Bosch_case_1.2.xlsx")
```

```
data = pd.concat([parte1, parte2], axis = 0)
data.head()
```

```
data.tail(1)
```

```
len(data)
```

Si procede ad eliminare tutte le osservazioni che hanno gli esiti 3 (Disabilitato) e 4 (Rilavorato), poichè a seguito della consulenza con il referente Astori di Bosch abbiamo ritenuto corretto ottenere il dataset pulito rispetto ai casi limite in cui l'osservazione del dataset non è in grado di riassumere l'intero corso degli eventi (es: il pezzo viene preso dagli scarti e rilavorato da una stazione in poi che non è quella iniziale)

```
data = data[(data.Esito_S10 != 3) &
     ((data.Esito_S20 != 3)&(data.Esito_S20 != 4)) &
     ((data.Esito_S40 != 3)&(data.Esito_S40 != 4)) &
     ((data.Esito_S50 != 3)&(data.Esito_S50 != 4)) &
     ((data.Esito_S60 != 3)&(data.Esito_S60 != 4))]
```

Si procede anche ad eliminare i casi "impossibili" in cui la variabile Esito_S10 assume valori 0, cioè i casi in cui il prodotto non viene nemmeno sottoposto alla lavorazione della stazione S10:

```
data = data[data.Esito_S10 != 0]
```

Si descrive brevemente il dataset

```
len(data)
```

```
data.dtypes
```

Dal colloquio avuto con il referente Astori è emerso che gli esiti delle stazioni di lavorazione S10, S20 e S40 sono poco affidabili in quanto vi sono casi in cui nonostante queste possano segnare esiti negativi della lavorazione (scarti) il prodotto finale è comunque conforme e convalidato.

###Nuove variabili

Nuova variabile date mesi/giorni

```
data["Data_Ingresso_day"] = data["Data_Ingresso"].apply(lambda x: x.date())
data["Data_Ingresso_day"].head()
```

```
data["Data_Ingresso_month"] = data["Data_Ingresso"].apply(lambda x: x.strftime("%Y-%m"))
data["Data_Ingresso_month"].head()
```

```
data["Data_Ingresso_year"] = data["Data_Ingresso"].apply(lambda x: x.strftime("%Y"))
data["Data_Ingresso_year"].head()
```

Nuova variabile "Success" rappresentativa dell'esito

```
data["Success"] = [1 if i == 70 else 0 for i in data.UltimaStazione ]
data.head()
```

Esistono dei casi in cui il valore del test finale svolto alla stazione S60 è negativo ma l'esito della produzione è positivo ed il prodotto viene considerato conforme (poichè raggiunge la stazione S70); Si procede ad eliminare tali osservazioni in quanto non vi è la certezza che siano conformi.

```
indice = data[(data.UltimaStazione == 70) & (data.Esito_S60 == 2)].index
data.drop(index = indice, inplace = True)
```

##Test di normalità della distribuzione delle variabili:

```
def normalita(serie):
  #Shapiro
  #H0 c'è normalità
  print(scipy.stats.shapiro(serie))
  if scipy.stats.shapiro(serie)[1]>0.05:
    print(str(serie.name) + " normalità per Shapiro")
  else:
    print(str(serie.name) + " NON normalità per Shapiro")
  print("\n")

  #Kolmogorov
  #H0 c'è normalità
  print(scipy.stats.kstest(serie, "norm"))
  if scipy.stats.kstest(serie, "norm")[1]>0.05:
    print(str(serie.name) + " normalità per Kolmogorov")
  else:
    print(str(serie.name) + " NON normalità per Kolmogorov")
  print("\n")

  #Skewness Test
  #H0 c'è normalità
  print(scipy.stats.normaltest(serie))
  if scipy.stats.normaltest(serie)[1]>0.05:
    print(str(serie.name) + " normalità per Skewness")
  else:
    print(str(serie.name) + " NON normalità per Skewness")
```

```
variabili = ['S10Coppia',
       'S10Angolo', 'S20Coppia', 'S20Angolo', 'S20Portata',
       'S40F2MomentoTorcMax', 'S40Vite1Coppia', 'S40Vite1Angolo',
       'S40Vite2Coppia', 'S40Vite2Angolo', 'S40Vite3Coppia', 'S40Vite3Angolo',
       'S50PressionePT', 'S50TenutaPZ', 'S60F2DepresMin', 'S60F2Coppia',
       'S60F2Velocita', 'S60F2TenutaVNR']

for i in variabili:
  print("\n")
  normalita(data[str(i)])
  print("\n")
  display("-"*100)
```
